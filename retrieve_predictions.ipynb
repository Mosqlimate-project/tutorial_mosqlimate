You can also retrieve the predictions saved in the platform. 


```python
import requests
```

It's just necessary to provide the `id` associated with the prediction. You can look at this number on the `predictions` page on the platform.

```python
id = 76  # from res
url = f"https://api.mosqlimate.org/api/registry/predictions/{id}"
res = requests.get(url)
res.json()
```

To access just the JSON associated with the prediction: 

```python
json_preds = res.json()['prediction']

json_preds
```

To transform the JSON in a pandas dataframe:

```python
import json
import pandas as pd
json_struct = json.loads(json_preds)    
df = pd.io.json.json_normalize(json_struct)
df.dates = pd.to_datetime(df.dates)
df.head()
```
